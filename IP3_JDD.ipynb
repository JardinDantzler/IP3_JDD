```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Step 1. Convert the Date to datatime and a column named "Company"

```python
def newdata(files):
    df = pd.read_csv(f"{files}.csv")
    df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
    df.insert(0, 'Company', files)
    return df
```

##### Step 2. Merge the data sets together 

```python
'Using just one line to merge them together before stacking'
df = pd.DataFrame().append([Hertz, Amprion, TenneTTSO, TransnetBW])
```

```python
df
```

```python
'Looking at the data by "Company" and "Year"'
df_year = df.groupby([df['Company'], df['Date'].dt.year]).mean()
```

```python
df_year
```

##### Step 3. Plotting Average amount of data for each company, in each year by "Noon" and "Midnight"

```python
Noon = pd.Series(df_year['12:00:00'])
Noon
```

```python
Midnight = pd.Series(df_year['00:00:00'])
Midnight
```

```python
'plotting the Average amount of data with our 2 new sets of data' 

figure1 = plt.figure(figsize=(10,5))
Noon.sort_values().plot(kind='barh', color='red')
plt.title('TW/hrs by Company by Year at Noon')
plt.ylabel('Company by year')
plt.savefig('noon_data.png', bbox_inches='tight')
```

```python
figure = plt.figure(figsize=(10,5))
Midnight.sort_values().plot(kind='barh', color='blue')
plt.title('TW/hrs by Company by Year at Midnight')
plt.ylabel('Company by year')
plt.savefig('midnight_data.png', bbox_inches='tight')
```

##### Step 4. Combining the two seperate graphs together into a single graph 

```python
figure2 = plt.figure(figsize=(10,7))

ax1 = figure2.add_subplot(211)
Midnight.sort_values().plot(kind='barh', color='silver')
plt.title('TW/hrs by Company by Year at Midnight')
plt.ylabel('Company by year')


ax2 = figure2.add_subplot(212, sharex=ax1)
Noon.sort_values().plot(kind='barh', color='gray')
plt.title('TW/hrs by Company by Year at Noon')
plt.ylabel('Company by year')

plt.savefig('combined_bar_chart.png', bbox_inches='tight')
```

##### Step 5. A function to show the data of a specific day for a specific company

```python
def daily_view(dataframe, company, date, color):
    figure3 = plt.figure(figsize = (10, 5))
    dataframe.query(f"Date == '{date}' and Company == '{company}'").iloc[:,2:].T.plot(kind='area', color = color)
    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.title(f'Power Production: {company}/{date}')
    plt.xlabel('Time of day')
    plt.ylabel('TW/hr Generated')
    return figure3
```

```python
daily_view(df, '50Hertz', '2019-08-24', 'purple')
```

#### Bonus 

```python
def daily_views(dataframe, companies, date,color):
    figure,axis=plt.subplots(len(companies))
    figure.tight_layout()
    for company in companies:
        company_data=dataframe.query(f'Company=="{company}" and Date=="{date}"')
        times=list(company_data.columns)[2:]
        axis[companies.index(company)].stackplot(range(len(times)),list(company_data.T[company_data.T.columns[0]])[2:],color=color[companies.index(company)])
        axis[companies.index(company)].set_title(f'Power Production:{date} for {company}')
        axis[companies.index(company)].set_xticks([0,20,40,60,80])
        axis[companies.index(company)].set_xticklabels(['00:00','05:00','10:00','15:00','20:00'])
daily_views(df,['50Hertz','Amprion','TransnetBW','TenneTTSO'],'2019-08-24',['Red','Green','Blue','purple'])
```
